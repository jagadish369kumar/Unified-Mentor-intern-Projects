```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\JAGADISH\Downloads\Bird Strikes data.xlsx - Bird Strikes.csv")
df
```

```python
df.columns
```

```python
# Display the first few rows of the dataframe to understand its structure
df.head()
```

```python
# Convert the FlightDate column to datetime format
df['FlightDate'] = pd.to_datetime(df['FlightDate'], errors='coerce')
```

```python
# Convert numeric columns to appropriate types
numeric_columns = ['Cost: Total $', 'Feet above ground', 'Number of people injured']
for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# Check for and handle missing values
missing_values_summary = df.isnull().sum()
```

```python
# Display the summary of missing values
missing_values_summary
```

```python
# Drop rows with critical missing values
critical_columns = ['FlightDate', 'Aircraft: Type', 'Airport: Name', 'When: Phase of flight']
df = df.dropna(subset=critical_columns)
```

```python
# Fill missing numeric values with zero or appropriate statistical measures
df['Cost: Total $'] = df['Cost: Total $'].fillna(0)
df['Feet above ground'] = df['Feet above ground'].fillna(0)
```

```python
# Verify if critical columns are now free of missing values
missing_values_summary_after_cleaning = df[critical_columns].isnull().sum()
```

```python
# Extract necessary columns for visualizations
columns_needed = [
    'Record ID', 'FlightDate', 'Aircraft: Type', 'Airport: Name', 'Altitude bin', 'Aircraft: Airline/Operator','Origin State', 
    'Effect: Impact to flight', 'Effect: Indicated Damage', 'When: Phase of flight', 'Cost: Total $', 
    'Feet above ground', 'Wildlife: Size', 'Wildlife: Species', 'Pilot warned of birds or wildlife?'
]

bird_strikes_cleaned_df = df[columns_needed]
```

```python
# Display the summary of missing values after cleaning
missing_values_summary_after_cleaning, bird_strikes_cleaned_df.head()
```
